```python
#imports das bibliotecas
import scipy.io as sio
import numpy as np
import os
from os import getcwd
from os.path import join
import matplotlib.pyplot as plt
from statsmodels.tsa.ar_model import AutoReg
from sklearn.linear_model import LogisticRegression
```

```python
fname = join(getcwd(),'data3SS2009.mat')

mat_contents = sio.loadmat(fname)

dataset = mat_contents['dataset']

N, Chno, Nc = dataset.shape
# N: number of samples
# Chno: number of channels
# Nc: number of cases

y = mat_contents['labels'].reshape(Nc)

y
y_target = y > 9 #criacao de uma matriz targets
y_target
print(y_target.shape)
print(y_target)
```

```python
#extração de caracteristicas
Na = 30
FeatArch2 = []
FeatArch3 = []
FeatArch4 = []
FeatArch5 = []
Ch1 = dataset[:,0,:] # célula de carga: força do shaker
Ch2 = dataset[:,1,:] # acelerômetro: base
Ch3 = dataset[:,2,:] # acelerômetro: 1o andar
Ch4 = dataset[:,3,:] # acelerômetro: 2o andar
Ch5 = dataset[:,4,:] # acelerômetro: 3o andar
for i in range(Nc):
  res1 = AutoReg(Ch2[:,i], lags=Na, trend='n').fit()
  res2 = AutoReg(Ch3[:,i], lags=Na, trend='n').fit()
  res3 = AutoReg(Ch4[:,i], lags=Na, trend='n').fit()
  res4 = AutoReg(Ch5[:,i], lags=Na, trend='n').fit()

  FeatArch2.append(res1.params)
  FeatArch3.append(res2.params)
  FeatArch4.append(res3.params)
  FeatArch5.append(res4.params)

FeatArch2 = np.array(FeatArch2)
FeatArch3 = np.array(FeatArch3)
FeatArch4 = np.array(FeatArch4)
FeatArch5 = np.array(FeatArch5)
```

```python
X1 = np.concatenate([FeatArch2,FeatArch3,FeatArch4,FeatArch5],axis=1)
X1.shape
```

```python
np.save("atv1_data_AR_appl.txt", X1)
```

```python
#Método para obter X2 PCA
from sklearn import decomposition
pca = decomposition.PCA(n_components=5)
pca.fit(X1)
X2 = pca.transform(X1)

a = pca.explained_variance_ratio_
sum = 0
for i in a:
  sum = sum + i

print("soma das variancias em cada coluna")
print(sum)
print("Formato da matriz X2")
print(X2.shape)
```

```python
#codigo para saber qual o numero de caracteristica minima que faz com que a variancia fique igual ou acima de 95%
from sklearn.decomposition import PCA
pca = decomposition.PCA()
pca.fit(X1)
cumsum = np.cumsum(pca.explained_variance_ratio_)
d = np.argmax(cumsum >= 0.95) + 1
print(d)
```

```python
#codigo direto para X2
pca = PCA(n_components = 0.95)
x_red = pca.fit_transform(X1)
x_red.shape
```

```python
#preparando datasets para teste e treino
"""
x = np.random.randint(0,849,size)
np.random.shuffle(x)
x.shape
training, test = x[:680,:], x[680:,:]
print("matriz de dados para treino")
print(training.shape)
print("matriz de dados para teste")
print(test.shape)
print(x)
"""
from sklearn.model_selection import train_test_split
X1_train,X1_test, Y1_train, Y1_test = train_test_split(X1,y_target,test_size=0.2,random_state=42)
X2_train,X2_test, Y2_train, Y2_test = train_test_split(X2,y_target,test_size=0.2,random_state=42)

print(X1_train.shape)
print(X2_train.shape)

```

```python
#realizando a escala dos dados de treino
from sklearn.preprocessing import MaxAbsScaler
transformer = MaxAbsScaler().fit(X1_train)
transformer.max_abs_ = 1
X1_train_scaled = transformer.transform(X1_train)
transformer2 = MaxAbsScaler().fit(X2_train)
transformer2.max_abs_ = 1
X2_train_scaled = transformer2.transform(X2_train)
```

```python
#plotando gráficos paralelo
from sklearn import preprocessing
import pandas as pd
scaler = preprocessing.MinMaxScaler(feature_range=(-1,1))
indDam = y>9
dfared = pd.concat([pd.DataFrame((scaler.fit_transform(X1))),pd.DataFrame({'target':indDam})], axis=1)

plt.figure(1)
pd.plotting.parallel_coordinates(dfared,'target',colormap=plt.get_cmap('Paired'))
plt.show()
indDam2 = y>9
dfared2 = pd.concat([pd.DataFrame((scaler.fit_transform(X2))),pd.DataFrame({'target':indDam2})], axis=1)

plt.figure(2)
pd.plotting.parallel_coordinates(dfared2,'target',colormap=plt.get_cmap('Paired'))
plt.show()
```

```python
#espaço para o codigo do modelo linear
%%timeit
softmax_reg_X1 = LogisticRegression(multi_class="multinomial", solver="lbfgs",C=10)
softmax_reg_X1.fit(X1_train_scaled,Y1_train)
```

```python
%%timeit
softmax_reg_X2 = LogisticRegression(multi_class="multinomial", solver="lbfgs",C=10)
softmax_reg_X2.fit(X2_train_scaled,Y2_train)
```

```python
#espaço para teste X1

Y1_predicted = softmax_reg_X1.predict(X1_test)
Y2_predicted = softmax_reg_X2.predict(X2_test)
```

```python
#matriz de confusão para os dados X1

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm = confusion_matrix(Y1_test,Y1_predicted)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)

disp.plot()
plt.show()
```

```python

cm2 = confusion_matrix(Y2_test,Y2_predicted)
disp2 = ConfusionMatrixDisplay(confusion_matrix=cm2)

disp2.plot()
plt.show()
```
